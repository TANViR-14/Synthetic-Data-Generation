```python
# Generate Bangladesh phone number
def generate_phone():
    return f"01{random.randint(3, 9)}{random.randint(10000000, 99999999)}"
```

```python
# Generate email from name
def generate_email(name):
    domains = ['gmail.com', 'yahoo.com', 'outlook.com']
    clean_name = name.lower().replace(' ', '_')
    return f"{clean_name}@{random.choice(domains)}"
```

```python
# 1. GENERATE DIM_SPECIALIZATION
NUM_SPECIALIZATIONS = 15
print("\nGenerating DIM_SPECIALIZATION table...")

specializations = [
    ('Cardiology', 'Rangpur Medical College', 2002),
    ('Dermatology', 'Dhaka Medical College', 1959),
    ('Pediatrics', 'Chittagong Medical College', 1984),
    ('Orthopedics', 'Sylhet MAG Osmani Medical College', 1966),
    ('Neurology', 'Rajshahi Medical College', 1957),
    ('Gynecology', 'Mymensingh Medical College', 1977),
    ('General Medicine', 'Rangpur Medical College', 2005),
    ('Ophthalmology', 'Dhaka Medical College', 2001),
    ('ENT', 'Khulna Medical College', 1991),
    ('Psychiatry', 'National Institute of Mental Health', 2012),
    ('Urology', 'Bangabandhu Sheikh Mujib Medical University', 2016),
    ('Gastroenterology', 'Dhaka Medical College', 1997),
    ('Nephrology', 'Kidney Foundation', 2011),
    ('Endocrinology', 'BIRDEM Hospital', 2008),
    ('Pulmonology', 'National Institute of Diseases of Chest', 2013)
]
```

```python
dim_specialization = pd.DataFrame({
    'SPECIALIZATION_ID': [f"S{i:03d}" for i in range(1, NUM_SPECIALIZATIONS + 1)],
    'SPECIALIZATION_TITLE': [s[0] for s in specializations],
    'INSTITUTE': [s[1] for s in specializations],
    'YEAR': [s[2] for s in specializations]
})

print(f"> Generated {len(dim_specialization)} specialization records")
dim_specialization
```

```python
# 2 GENERATE DIM_DOCTOR
NUM_DOCTORS = 150
print("\nGenerating DIM_DOCTOR table...")

first_names = ['Ahmed', 'Fatima', 'Musa', 'Ayesha', 'Karim', 'Nusrat', 'Hasan', 'Sultana', 
               'Rahim', 'Jannat', 'Nakib', 'Razia', 'Kabir', 'Nasrin', 'Farhan']
last_names = ['Khan', 'Chowdhury', 'Islam', 'Rahman', 'Akhter']
```

```python
doctor_data = []
for i in range(1, NUM_DOCTORS + 1):
    name = f"Dr. {random.choice(first_names)} {random.choice(last_names)}"
    gender = random.choice(['Male', 'Female'])
    
    # Pick specialization ID as string matching dim_specialization
    spec_id = random.choice(dim_specialization['SPECIALIZATION_ID'].tolist())
    
    qualification = random.choice(['MBBS, FCPS', 'MBBS, MD', 'MBBS, MS', 'MBBS, FRCS'])
    experience = random.randint(2, 25)
    
    # Get the title from dim_specialization
    specialization_title = dim_specialization.loc[
        dim_specialization['SPECIALIZATION_ID'] == spec_id, 'SPECIALIZATION_TITLE'
    ].values[0]
    
    doctor_data.append({
        'DOC_ID': f"D{i:03d}",
        'SPECIALIZATION_ID': spec_id,
        'DOCTOR_NAME': name,
        'GENDER': gender,
        'SPECIALIZATION': specialization_title,
        'QUALIFICATION': qualification,
        'YEAR_OF_EXPERIENCE': experience,
        'PHONE_NO': generate_phone(),
        'EMAIL': generate_email(name)
    })

dim_doctor = pd.DataFrame(doctor_data)


print(f"> Generated {len(dim_doctor)} doctor records")
dim_doctor
```

```python
# 4. GENERATE DIM_PATIENT
NUM_PATIENTS = 500
print("\nGenerating DIM_PATIENT table...")

patient_first_names = ['Rahim', 'Karim', 'Fatema', 'Ayesha', 'Salam', 'Nazia', 'Rakib', 
                       'Shirin', 'Jahid', 'Nasima', 'Rafiq', 'Selina', 'Milon', 'Rupa']
patient_last_names = ['Uddin', 'Akter', 'Begum', 'Mia', 'Sheikh', 'Sarkar', 'Mondal', 
                      'Haque', 'Biswas', 'Talukdar', 'Paul', 'Barua']

medical_conditions = ['Hypertension', 'Diabetes', 'Asthma', 'None', 'Allergy', 'Arthritis']
```

```python
patient_data = []
for i in range(1, NUM_PATIENTS + 1):
    name = f"{random.choice(patient_first_names)} {random.choice(patient_last_names)}"
    age = random.randint(18, 80)
    gender = random.choice(['Male', 'Female'])
    location_id = random.choice(dim_location['LOCATION_ID'].tolist())

    
    patient_data.append({
        'P_ID': f"P{i:04d}",
        'LOCATION_ID': location_id,
        'PATIENT_NAME': name,
        'AGE': age,
        'GENDER': gender,
        'PHONE_NO': generate_phone(),
        'EMAIL': generate_email(name),
        'MEDICAL_HISTORY': random.choice(medical_conditions)
    })

dim_patient = pd.DataFrame(patient_data)

print(f"> Generated {len(dim_patient)} patient records")
dim_patient
```

```python
# 4. GENERATE DIM_HEALTHCARE
NUM_HEALTHCARE = 80

print("\nGenerating DIM_HEALTHCARE table...")

healthcare_names = ['City Hospital', 'General Clinic', 'Medical Center', 'Health Complex', 
                    'Diagnostic Center', 'Poly Clinic', 'Specialized Hospital']
healthcare_types = ['Hospital', 'Clinic', 'Diagnostic Center', 'Health Complex']
branches = ['Main Branch', 'Branch A', 'Branch B', 'Branch C']
```

```python
healthcare_data = []
for i in range(1, NUM_HEALTHCARE + 1):
    location_id = random.choice(dim_location['LOCATION_ID'].tolist())
    location_name = dim_location[dim_location['LOCATION_ID'] == location_id]['UPAZILLA'].values[0]
    
    healthcare_data.append({
        'HEALTHCARE_ID': f"H{i:03d}", 
        'LOCATION_ID': location_id,
        'HEALTHCARE_NAME': f"{random.choice(healthcare_names)} {location_name}",
        'HEALTHCARE_CODE': f"HC{str(i).zfill(4)}",
        'BRANCH_NAME': random.choice(branches),
        'BRANCH_CODE': f"BR{random.randint(100, 999)}",
        'HEALTHCARE_TYPE': random.choice(healthcare_types)
    })

dim_healthcare = pd.DataFrame(healthcare_data)

print(f"> Generated {len(dim_healthcare)} healthcare facility records")
dim_healthcare
```

```python
# 5. GENERATE DIM_TEST
NUM_TESTS = 200
print("\nGenerating DIM_TEST table...")

tests = [
    ('Blood Test', 'Blood', 'CBC, Blood Sugar'),
    ('X-Ray', 'Image', 'Chest, Bone'),
    ('ECG', 'Heart', 'Cardiac activity'),
    ('Ultrasound', 'Image', 'Abdominal scan'),
    ('MRI', 'Image', 'Brain, Spine'),
    ('CT Scan', 'Image', 'Full body'),
    ('Urine Test', 'Urine', 'Routine examination'),
    ('Liver Function Test', 'Blood', 'Hepatic panel'),
    ('Kidney Function Test', 'Blood', 'Renal panel'),
    ('Thyroid Test', 'Blood', 'TSH, T3, T4'),
    ('Blood Pressure', 'Vital', 'BP measurement'),
    ('Diabetes Test', 'Blood', 'HbA1c, Fasting'),
    ('Cholesterol Test', 'Blood', 'Lipid profile'),
    ('COVID-19 Test', 'Swab', 'RT-PCR'),
    ('Stool Test', 'Stool', 'Routine examination'),
    ('Biopsy', 'Tissue', 'Histopathology'),
    ('Endoscopy', 'Image', 'GI tract'),
    ('Colonoscopy', 'Image', 'Colon examination'),
    ('Mammography', 'Image', 'Breast scan'),
    ('Bone Density', 'Image', 'DEXA scan'),
    ('Allergy Test', 'Blood', 'IgE levels'),
    ('HIV Test', 'Blood', 'Antibody detection'),
    ('Pregnancy Test', 'Urine', 'hCG detection'),
    ('Pap Smear', 'Cell', 'Cervical screening'),
    ('Semen Analysis', 'Semen', 'Fertility test')
]
```

```python
test_data = []
for i in range(1, NUM_TESTS + 1):
    healthcare_id = random.choice(dim_healthcare['HEALTHCARE_ID'].tolist())
    test_info = tests[i-1] if i-1 < len(tests) else random.choice(tests)
    
    test_data.append({
        'TEST_ID': f"T{i:05d}", 
        'HEALTHCARE_ID': healthcare_id,
        'TEST_NAME': test_info[0],
        'SAMPLE_TYPE': test_info[1],
        'REMARKS': test_info[2]
    })

dim_test = pd.DataFrame(test_data)

print(f"> Generated {len(dim_test)} test records")
dim_test
```

```python
# 6. GENERATE DIM_TIME
import calendar

print("\nGenerating DIM_TIME table...")

print("\nGenerating DIM_TIME table...")

time_data = []

# Define years to include
years = [2019, 2020]

for year in years:
    for month in range(1, 13):
        # Get number of days in this month (handles leap years)
        _, num_days = calendar.monthrange(year, month)

        for day in range(1, num_days + 1):
            for hour in range(24):
                for minute in [0, 15, 30, 45]:
                    # Create TIME_ID as string: DDMMYYYY
                    time_id = f"{year}{month:02d}{day:02d}"

                    time_data.append({
                        'TIME_ID': time_id,
                        'DAY': day,
                        'MONTH': month,
                        'YEAR': year,
                        'HOUR': hour,
                        'MINUTE': minute
                    })

dim_time = pd.DataFrame(time_data)

print(f"> Generated {len(dim_time)} time slot records")
dim_time
```

```python
# 7. GENERATE DIM_CONSULTATION
NUM_CONSULTATIONS = 300
print("\nGenerating DIM_CONSULTATION table...")

consultation_data = []
for i in range(1, NUM_CONSULTATIONS + 1):
    spec_id = random.randint(1, NUM_SPECIALIZATIONS)
    
    consultation_data.append({
        'CONSUL_ID': f"T{i:03d}",
        'SPECIALIZATION_ID': random.choice(dim_specialization['SPECIALIZATION_ID'].tolist()),
        'DURATION': random.choice([15, 30, 45, 60]),  # minutes
        'FEE': random.choice([300, 500, 700, 1000, 1500])  # BDT
        #'RATINGS': round(random.uniform(3.5, 5.0), 1)
    })

dim_consultation = pd.DataFrame(consultation_data)

print(f"> Generated {len(dim_consultation)} consultation type records")
dim_consultation
```

```python
# 8. GENERATE FACT_TELEMEDICINE (using JOIN operations)
NUM_TELEMEDICINE_RECORDS = 1000
print("\nGenerating FACT_TELEMEDICINE table...")
print("This table joins data from all dimension tables...")

telemedicine_data = []

for i in range(1, NUM_TELEMEDICINE_RECORDS + 1):
    # Randomly select a patient
    patient_id = random.choice(dim_patient['P_ID'].tolist())
    patient_row = dim_patient[dim_patient['P_ID'] == patient_id].iloc[0]
    
    # Select a healthcare facility (prefer same location as patient)
    if random.random() < 0.7:  # 70% same location
        healthcare_row = dim_healthcare[dim_healthcare['LOCATION_ID'] == patient_row['LOCATION_ID']]
        if len(healthcare_row) > 0:
            healthcare_row = healthcare_row.sample(1).iloc[0]
        else:
            healthcare_row = dim_healthcare.sample(1).iloc[0]
    else:
        healthcare_row = dim_healthcare.sample(1).iloc[0]
    
    healthcare_id = healthcare_row['HEALTHCARE_ID']
    
    # Select a consultation type
    consul_row = dim_consultation.sample(1).iloc[0]
    consul_id = consul_row['CONSUL_ID']
    
    # Select a doctor with matching specialization
    matching_doctors = dim_doctor[dim_doctor['SPECIALIZATION_ID'] == consul_row['SPECIALIZATION_ID']]
    if len(matching_doctors) > 0:
        doc_row = matching_doctors.sample(1).iloc[0]
    else:
        doc_row = dim_doctor.sample(1).iloc[0]
    doc_id = doc_row['DOC_ID']
    
    # Select a test from the same healthcare facility
    matching_tests = dim_test[dim_test['HEALTHCARE_ID'] == healthcare_id]
    if len(matching_tests) > 0 and random.random() < 0.6:  # 60% chance of test
        test_row = matching_tests.sample(1).iloc[0]
        test_id = test_row['TEST_ID']
    else:
        test_id = None  # No test prescribed
    
    # Select a time slot
    time_id = random.choice(dim_time['TIME_ID'].tolist())
    
    # Generate consultation details
    conditions = ['Fever', 'Headache', 'Cough', 'Back Pain', 'Stomach Pain', 
                  'High BP', 'Diabetes Check', 'General Checkup', 'Follow-up']
    
    medicines = ['Paracetamol 500mg', 'Amoxicillin 250mg', 'Omeprazole 20mg', 
                 'Metformin 500mg', 'None prescribed', 'Cetirizine 10mg']
    
    instructions = ['Take after meals', 'Take before sleep', 'Complete the course', 
                    'Follow up after 1 week', 'Rest and drink plenty of water']
    
    feedbacks = ['Very helpful', 'Good consultation', 'Satisfied', 'Excellent service', 
                 'Doctor was patient', 'Quick diagnosis']
    
    telemedicine_data.append({
        'TM_ID': f"M{i:05d}",
        'HEALTHCARE_ID': healthcare_id,
        'P_ID': patient_id,
        'CONSUL_ID': consul_id,
        'DOC_ID': doc_id,
        'TEST_ID': test_id,
        'TIME_ID': time_id,
        'CONDITION': random.choice(conditions),
        'MEDICINE': random.choice(medicines),
        'INSTRUCTIONS': random.choice(instructions),
        'FEEDBACK': random.choice(feedbacks)
    })

fact_telemedicine = pd.DataFrame(telemedicine_data)

print(f"> Generated {len(fact_telemedicine)} telemedicine consultation records")
fact_telemedicine 
```

```python
# 9. SAVE ALL TABLES TO CSV FILES
print("\n" + "="*50)
print("SAVING ALL TABLES TO CSV FILES")
print("="*50)

dim_specialization.to_csv('dim_specialization.csv', index=False)
print("> Saved: dim_specialization.csv")

dim_doctor.to_csv('dim_doctor.csv', index=False)
print("> Saved: dim_doctor.csv")

dim_patient.to_csv('dim_patient.csv', index=False)
print("> Saved: dim_patient.csv")

dim_healthcare.to_csv('dim_healthcare.csv', index=False)
print("> Saved: dim_healthcare.csv")

dim_test.to_csv('dim_test.csv', index=False)
print("> Saved: dim_test.csv")

dim_time.to_csv('dim_time.csv', index=False)
print("> Saved: dim_time.csv")

dim_consultation.to_csv('dim_consultation.csv', index=False)
print("> Saved: dim_consultation.csv")

fact_telemedicine.to_csv('fact_telemedicine.csv', index=False)
print("> Saved: fact_telemedicine.csv")
```

```python
# ============================================================================
# STEP 2: MERGE DATA (Simple left joins)
# ============================================================================
print("\n[Step 2] Merging data...")

# Start with fact table
merged = fact_telemedicine.copy()

# Merge with patient (get patient info)
merged = merged.merge(dim_patient, on='P_ID', how='left')

# Merge with doctor (get doctor info)
merged = merged.merge(dim_doctor, on='DOC_ID', how='left')

# Merge with healthcare (get facility info)
merged = merged.merge(dim_healthcare, on='HEALTHCARE_ID', how='left')

# Merge with consultation (get duration and fee)
merged = merged.merge(dim_consultation, on='CONSUL_ID', how='left')

# Merge with time (get year, month, day)
merged = merged.merge(dim_time[['TIME_ID', 'YEAR', 'MONTH', 'DAY', 'HOUR']].drop_duplicates(), 
                      on='TIME_ID', how='left')

# Merge with location (get district and upazilla)
merged = merged.merge(dim_location, 
                      left_on='LOCATION_ID_x', 
                      right_on='LOCATION_ID', 
                      how='left')

print(f"✓ Merged dataset created with {len(merged)} records")
merged
```

```python
# TABLE 5.3: OLAP ROLL-UP Operations

print("\n" + "="*70)
print("TABLE 5.3: ROLL-UP Operations - Temporal Aggregation")
print("="*70)

# Yearly aggregation
yearly = merged.groupby('YEAR').agg({
    'TM_ID': 'count',
    'FEE': 'mean'
}).reset_index()
yearly.columns = ['Year', 'Total Consultations', 'Average Fee (BDT)']
yearly['Average Fee (BDT)'] = yearly['Average Fee (BDT)'].round(0)

print("\nYearly Summary:")
print(yearly.to_string(index=False))

# Monthly breakdown for first year
print(f"\nMonthly Breakdown (Year {yearly['Year'].iloc[0]}):")
year1 = merged[merged['YEAR'] == yearly['Year'].iloc[0]]
monthly = year1.groupby('MONTH').agg({
    'TM_ID': 'count',
    'FEE': 'mean'
}).reset_index()
monthly.columns = ['Month', 'Consultations', 'Avg Fee (BDT)']
monthly['Avg Fee (BDT)'] = monthly['Avg Fee (BDT)'].round(0)
print(monthly.head(6).to_string(index=False))
```

```python
# ============================================================================
# TABLE 5.4: Geographic ROLL-UP
# ============================================================================
print("\n" + "="*70)
print("TABLE 5.4: Geographic ROLL-UP - District Level")
print("="*70)

# Group by district
geo = merged.groupby('DISTRICT').agg({
    'TM_ID': 'count',
    'P_ID': 'nunique'
}).reset_index()
geo.columns = ['District', 'Total Consultations', 'Unique Patients']
geo['Consultations per Patient'] = (geo['Total Consultations'] / geo['Unique Patients']).round(2)
geo = geo.sort_values('Total Consultations', ascending=False).head(10)

print(geo.to_string(index=False))
```

```python
# TABLE 5.5: DICE Operation (Multi-dimensional filtering)

print("\n" + "="*70)
print("TABLE 5.5: DICE Operation - Filtered Analysis")
print("="*70)

# Get top specialty and top district
top_spec = merged['SPECIALIZATION'].value_counts().index[0]
top_district = merged['DISTRICT'].value_counts().index[0]
year_filter = merged['YEAR'].mode()[0]

# Apply DICE filter
dice_data = merged[
    (merged['SPECIALIZATION'] == top_spec) &
    (merged['DISTRICT'] == top_district) &
    (merged['YEAR'] == year_filter)
]

print(f"\nFilter Applied:")
print(f"  - Specialization: {top_spec}")
print(f"  - District: {top_district}")
print(f"  - Year: {year_filter}")
print(f"\nTotal records after DICE: {len(dice_data)}")

if len(dice_data) > 0:
    # Summarize by healthcare type
    dice_summary = dice_data.groupby('HEALTHCARE_TYPE').agg({
        'TM_ID': 'count',
        'FEE': 'mean',
        'DURATION': 'mean'
    }).reset_index()
    dice_summary.columns = ['Healthcare Type', 'Consultations', 'Avg Fee (BDT)', 'Avg Duration (min)']
    dice_summary['Avg Fee (BDT)'] = dice_summary['Avg Fee (BDT)'].round(0)
    dice_summary['Avg Duration (min)'] = dice_summary['Avg Duration (min)'].round(1)
    
    print("\nDICE Results:")
    print(dice_summary.to_string(index=False))
```
